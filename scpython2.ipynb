```python
"""
SECCION LISTAS Y TUPLAS
"""
frutas = ['4','peras','naranjas']
print(f"Lista original: {frutas}")

frutas.append('fresa') #usamos append si queremos añadir solo un objeto, así sea una lista compleja
print(f"Lista 2 : {frutas}")

frutas.insert(2 , 'uvas') #insertamos por posición
print(f"Lista 3 : {frutas}")

print(f"i = 0 : {frutas[0]}")
print(f"i = 1 : {frutas[1]}")
print(f"i = 2 : {frutas[2]}")
print(f"i = 3 : {frutas[3]}")
print(f"i = 4 : {frutas[4]}")

frutas.extend(['kiwi', 'mango']) #usamos extend si queremos fusionar dos listas  en una sola
print(f"Lista 4 : {frutas}")

frutaEliminada = frutas.pop(2)
print(f"Fruta eliminada: {frutaEliminada}")
print(f"Lista 5 : {frutas}")

frutas.remove('naranjas') #usamos remove si sabemos qué elemento queremos eliminar
print(f"Lista 6 : {frutas}")

frutas.reverse()
print(f"Lista 7 : {frutas}") #reverse invierte la lista

frutas.sort()
print(f"Lista 8 : {frutas}") #sort ordena la lista siempre y cuando sea solo  de strings

tupla_demo = (2,[3,45],'Alexander') #una tupla es una estructura inmutable
print(tupla_demo)
var1, var2,var3= tupla_demo #desarticulamos la tupla para poder imprimirlos
print(var1)
print(var2)
print(var3)
```
